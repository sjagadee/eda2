### Load Fifa Data

```python
fifa = pd.read_csv('/content/drive/MyDrive/DataAnalysis/Data/fifa_data.csv')
fifa.head()
```

### Histogram

```python
bins = [i for i in range(30, 110, 10)]

plt.hist(fifa.Overall, bins=bins, color='#22bfb4')

plt.xticks(bins)
plt.title('Distribution of Player Ratings', fontname='cmb10', fontsize= 20)
plt.xlabel('Skill Level', fontname='cmb10', fontsize= 14)
plt.ylabel('Number of Players', fontname='cmb10', fontsize= 14)

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
left, right
```

```python
plt.pie(
    [left, right],
    labels=['Left', 'Right'],
    colors=['#a275d6', '#22bf66'],
    autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players', fontname='cmb10', fontsize= 18)

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
light = fifa.loc[fifa['Weight'] <= 125].count()[0]
light_medium = fifa.loc[(fifa['Weight'] > 125) & (fifa['Weight'] <= 150)].count()[0]
medium = fifa.loc[(fifa['Weight'] > 150) & (fifa['Weight'] <= 175)].count()[0]
heavy_medium = fifa.loc[(fifa['Weight'] > 175) & (fifa['Weight'] <= 200)].count()[0]
heavy = fifa.loc[fifa['Weight'] > 200].count()[0]
```

```python
weights = [light, light_medium, medium, heavy_medium, heavy]
labels = ['Light (Under 125)', 'Light Medium (125-150)', 'Medium (150-175)', 'Heavy Medium (175-200)', 'Heavy (Over 200)']
explode = [0.4, 0.2, 0, 0, 0.3]

plt.style.use('ggplot')

plt.pie(
    weights,
    labels=labels,
    # colors=['#75d6ba', '#7584d6', '#ae75d6', '#d67975', '#d13c46'],
    autopct='%.2f %%', pctdistance=0.8,
    explode=explode)

plt.title('Weight Distribution of FIFA Players (in lbs)', fontname='cmb10', fontsize= 18)

plt.show()
```

### Box and Wiskers

```python
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
revs = fifa.loc[fifa['Club'] == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

plt.style.use('default')


boxes = plt.boxplot(
    [barcelona, madrid, revs],
    tick_labels=labels,
    patch_artist=True)

for bax in boxes['boxes']:
    # edge color
    bax.set(color='#3c86d1', linewidth=2)

    # fill color
    bax.set(facecolor='#75d6ba')

plt.title('FIFA Clubs Comparision', fontname='cmb10', fontsize= 18)
plt.xlabel('Clubs', fontname='cmb10', fontsize= 14)
plt.ylabel('Overall Rating', fontname='cmb10', fontsize= 14)

plt.show()
```
