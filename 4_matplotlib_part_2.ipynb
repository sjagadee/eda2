## Real World Examples

### Line Graph - Gas Prices data

```python
gas = pd.read_csv('/content/drive/MyDrive/DataAnalysis/Data/gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,6))

# plt.plot(gas['Year'], gas['USA'], 'b.-', label='USA')
# plt.plot(gas['Year'], gas['Canada'], 'r.-', label='Canada')
# plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')
# plt.plot(gas['Year'], gas['Australia'], 'y.-', label='Australia')

#another way to do
lookat_countries = ['USA', 'Canada', 'South Korea', 'Australia']

for country in gas.columns[1:]:
    if country in lookat_countries:
        plt.plot(gas['Year'], gas[country], marker='.', label=country)

plt.title('Gas Prices over the years (Per Gallon)', fontname='cmb10', fontsize= 20)
plt.xlabel('Year', fontname='Liberation Mono', fontsize= 16)
plt.ylabel('Price in USD ($)', fontname='Liberation Mono', fontsize= 16)
plt.legend()

plt.xticks(gas['Year'][::3].tolist())

# to save use - plt.savefig('to_location/image_name.png', dpi=300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('/content/drive/MyDrive/DataAnalysis/Data/fifa_data.csv')
fifa.head()
```

### Histogram

```python
bins = [i for i in range(30, 110, 10)]

plt.hist(fifa.Overall, bins=bins, color='#22bfb4')

plt.xticks(bins)
plt.title('Distribution of Player Ratings', fontname='cmb10', fontsize= 20)
plt.xlabel('Skill Level', fontname='cmb10', fontsize= 14)
plt.ylabel('Number of Players', fontname='cmb10', fontsize= 14)

plt.show()
```
