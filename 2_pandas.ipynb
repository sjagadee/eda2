## Making changes to the data

```python
# creating total column by adding rows manually
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
df.head()
```

```python
# drop column

df = df.drop(columns=['Total'])
df.head()
```

```python
# add columns 4 to 9 elements using iloc and creating a new total column
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head()
```

```python
# Reorder columsn

cols = list(df.columns.values) # getting all the columns in list

# rearranging the columns accordingly - we can only concatinate lists (so all the columns should be list)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

```python
df.head()
```

## Filtering Data

```python
# filtering using multiple condition (and  = &, or = |) using location (loc)
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["Attack"] > 60)]
new_df
```

```python
# if we see the old index is still coming through - so we can reset index
# new_df = new_df.reset_index()
# new_df

# but we have created a column with index to remove the same
new_df = new_df.reset_index(drop=True)
new_df

# we do not have to save it back
# new_df.reset_index(drop=True, inplace=True)
```

```python
# advanced filtering for text haveing "Mega"

# 1. get all the pokemon with mega
df.loc[df['Name'].str.contains('Mega')]
```

```python
# to do not of the above list

df.loc[~df['Name'].str.contains('Mega')]
```

```python
# We can not only do string contains as shown in the above example. We can also use Regex filtering
import re

df.loc[df["Type 1"].str.contains("Fire|Grass", regex=True)]
```

```python
# to ignore case

df.loc[df["Type 1"].str.contains("fire|grass", flags = re.I, regex=True)]
```

```python
# example to get all pokemons to start with pi

df.loc[df["Name"].str.contains("^pi[a-zA-Z ]", flags = re.I, regex=True)]
```

## Conditional Changes

```python
# to change column based on condition and after comma we need to mention the name of the column
df.loc[df["Type 1"] == "Flamer", "Type 1"] = "Fire"

# df.loc[df["Type 1"] == "Fire", "Legendary"] = True
```

## Aggregate Statistics (Groupby)

```python
df.head()
```

```python
# groupby type 1 and get a mean for a specific column

df.groupby(["Type 1"])["Attack"].mean()
```

```python
# groupby type 1 and get a mean for all the columns

# step 1 = get all the list of columns that are numaricals
df_grouped = df.columns[df.dtypes == 'int64']
columns_with_nums = df_grouped.to_list()
print(columns_with_nums)

# step 2 = use this list and pass it to the grouped dataframe for all the required columns
# need to pass the list enclosed in square brackets [['Attack', 'Defense']]

df.groupby(["Type 1"])[columns_with_nums].mean()
```

```python
df.groupby(["Type 1"])[columns_with_nums].mean().sort_values('Defense', ascending=False)
```

```python
df.groupby(["Type 1"])[columns_with_nums].sum()
df.groupby(["Type 1"])[columns_with_nums].count()
```

```python
# group by multiple column names
df.groupby(["Type 1", "Type 2"])[columns_with_nums].count()['Total']
```
